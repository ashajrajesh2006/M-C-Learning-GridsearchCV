```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first='True')
```

```python
dataset
```

```python
dataset.columns
```

```python
independant=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
independant
```

```python
dependant=dataset[['charges']]
```

```python
dependant
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'criterion': ['squared_error', 'absolute_error'],
              'max_features': [None, 'sqrt', 'log2'],
              'n_estimators': [10, 100]}
rf = RandomForestRegressor(random_state=42)

grid = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=5,
    refit=True,
    verbose=3,
    n_jobs=-1
    )

grid.fit(independant, dependant)
```

```python
GridSearchCV(cv=5, error_score='raise',
             estimator=RandomForestRegressor(criterion='squared_error', max_depth=None,
                                             max_features=None,
                                             max_leaf_nodes=None,
                                             min_impurity_decrease=0.0,
                                             min_samples_leaf=1,
                                             min_samples_split=2,
                                             min_weight_fraction_leaf=0.0,
                                             random_state=None,
                                             n_estimators=rf), n_jobs=-1,
             param_grid={'criterion': ['squared_error', 'absolute_error'],'max_features': [None, 'sqrt', 'log2'],'n_estimators': [100, 200]},
             pre_dispatch='2*n_jobs', refit=True, return_train_score=False,scoring=None, verbose=3)
```

```python
# print best parameter after tuning
#print(grid.best_params_)
re=grid.cv_results_ #print(re)

print("The R_score value for best parameter {}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
```
